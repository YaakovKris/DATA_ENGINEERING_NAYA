## Step 1 — Clean the data
write a function the recives a `list` of `dict` - assume each dict have the exact same keys.

assume all values are always string

and returns a new cleand list.


For every order:

  * remove extra spaces from customer names
  * make customer names title case
  * convert the price to an integer
  * remove rows with negative price
  * remove rows with missing price

Example output:
```python
[
    {"customer": "Alice", "product": "Laptop", "price": 1200},
    ...
]

```

````python
orders =  [
    {"customer": " Alice ", "product": "Laptop", "price": "1200"},
    {"customer": "bot", "product": "Mouse", "price": "-999"},
    {"customer": "bob", "product": "Mouse", "price": "25"},
    {"customer": "ALICE", "product": "Keyboard", "price": "80"},
    {"customer": "Bob ", "product": "Monitor", "price": "300"},
    {"customer": "tstAcnt", "product": "Mouse", "price": ""},
    {"customer": " Charlie", "product": "USB Cable", "price": "15"},
]
````

````python
def clean_orders(orders):
    cleaned = []

    for order in orders:
        customer = order["customer"].strip().title()
        product = order["product"]
        price = order["price"]

        if price == "":
            continue

        price = int(price)

        if price < 0:
            continue

        cleaned.append({
            "customer": customer,
            "product": product,
            "price": price
        })

    return cleaned
````

````python
def generate_report(summary):
    report = []

    for customer in summary:
        total_orders = summary[customer]["total_orders"]
        total_spent = summary[customer]["total_spent"]

        report.append(
            f"{customer} placed {total_orders} orders and spent ${total_spent}"
        )

    return report
````
